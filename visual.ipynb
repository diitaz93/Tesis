```python
data=genfromtxt('un.dat',delimiter="\t")
N=len(data[0,:])-1 #Se ignora la ultima entrada ya que es un NaN
T=len(data[:,0])/3
print N
print T
```

```python
C=range(1,N+1)
for i in xrange(T):
    x=data[3*i,:N]
    y=data[3*i+1,:N]
    a=data[3*i+2,:N]
    U=cos(a)
    V=sin(a)
    fig, ax = subplots()
    ax.quiver(x,y,U,V,C)
    ax.axis([-0.25,1.25,-0.25,1.25])
```

```python
x0=data[0,:N]
y0=data[1,:N]
a0=data[2,:N]
xf=data[3,:N]
yf=data[4,:N]
af=data[5,:N]
#print x0
#print y0
#print a0
#print xf
#print yf
print af
```
