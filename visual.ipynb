```python
data=genfromtxt('un.dat',delimiter="\t")
N=len(data[0,:])-1 #Se ignora la ultima entrada ya que es un NaN
T=len(data[:,0])/3
print N
print T
```

```python
C=range(1,N+1)

for i in xrange(T):
    x=data[3*i,:N]
    y=data[3*i+1,:N]
    a=data[3*i+2,:N]
    U=cos(a)
    V=sin(a)
    fig, ax = subplots()
    ax.quiver(x,y,U,V,C)
    ax.axis([-0.25,1.25,-0.25,1.25])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'ro', animated=True)

def init():
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    return ln,

def update(frame):
    xdata.append(frame)
    ydata.append(np.sin(frame))
    ln.set_data(xdata, ydata)
    return ln,

ani = FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 128),
                    init_func=init, blit=True)
plt.show()
```
