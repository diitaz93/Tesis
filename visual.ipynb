```python
data=genfromtxt('un.dat',delimiter="\t")
N=len(data[0,:])-1
N
```

```python
x0=data[0,:N]
y0=data[1,:N]
a0=data[2,:N]
xf=data[3,:N]
yf=data[4,:N]
af=data[5,:N]
#print x0
#print y0
#print a0
#print xf
#print yf
print af
```

```python
U0=cos(a0)
V0=sin(a0)
Uf=cos(af)
Vf=sin(af)
C=range(1,N+1)

fig, ax = subplots()
ax.quiver(x0,y0,U0,V0,C)
ax.axis([-0.5,1.5,-0.5,1.5])
fig, ax = subplots()
ax.quiver(xf,yf,Uf,Vf,C)
ax.axis([-0.5,1.5,-0.5,1.5])
```
