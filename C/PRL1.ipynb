# Para las siguientes figuras, $N=300$, $v=0.03$, $r=1$, y el número de pasos es 20

## Figura 1a: Inicialización. $L=7$, $\eta=2.0$
20 pasos de tiempo

```python
data=genfromtxt('un.dat',delimiter="\t")
N=len(data[0,:])-1 #Se ignora la ultima entrada ya que es un NaN
T=len(data[:,0])/3
x0=data[0,:]
y0=data[1,:]
vx=cos(data[2,:])
vy=sin(data[2,:])
print N
print T
figure (figsize=[12,12])
quiver(x0,y0,vx,vy)
show()
```

## Figura 1b: $L=25$, $\eta=0.1$

```python
data=genfromtxt('un.dat',delimiter="\t")
N=len(data[0,:])-1 #Se ignora la ultima entrada ya que es un NaN
T=len(data[:,0])/3
print N
print T
figure(figsize=[12,12])
for i in range(1,T):
    x=data[3*i,:N]
    y=data[3*i+1,:N]
    scatter(x,y,s=2)#,c=t)
    hold
scatter(x,y,s=30, marker='^',c='red')
show()
```
